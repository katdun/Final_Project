```python
#get the number of movies/shows that are rated 
#ie it excludes any title that has NaN as a rating
def nancount(service_imdb_score):
    count = 0
    for x in service_imdb_score:
        if 0<x<10:
            count+=1
    return count
        
```

```python
def getgenrecount(genre_list):
    comedy = 0
    romance = 0
    drama = 0
    action =0
    family = 0
    documentation = 0
    other = 0

    for x in genre_list:
        #print(x)
        if 'comedy' in x:
            comedy +=1
        if 'romance' in x:
            romance+=1
        if 'drama' in x:
            drama+= 1
        if 'action' in x:
            action+= 1
        if 'family' in x:
            family +=1
        if 'documentation' in x:
            documentation += 1
        else :
            other += 1
    
    return {"comedy" : comedy, "romance" : romance, "drama" : drama, "action" : action, 
            "family": family, "documentary": documentation, "other" : other}
```

```python
netflix_top_genre_count = getgenrecount(netflix_genre)
hulu_top_genre_count = getgenrecount(hulu_genre)
amazon_top_genre_count = getgenrecount(amazon_genre)
disney_top_genre_count = getgenrecount(disney_genre)
HBO_top_genre_count = getgenrecount(HBO_genre)
```
